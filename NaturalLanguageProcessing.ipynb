# **Stemming**
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or
to the roots of words known as a lemma. Stemming is important in natural language understanding
(NLU) and natural language processing (NLP).

```python
from nltk.stem import PorterStemmer
```

```python
words = ["running","playing","walking","jumping","reading","writing",
"swimming","driving","eating","sleeping","played","walked","jumped",
"cooked","cleaned","watched","studied","worked","helped","talked",
"quickly","slowly","happily","sadly","easily","strongly","clearly",
"carefully","loudly","softly","cats","dogs","boxes","classes","buses",
"cars","trees","books","tables","chairs"]
```

```python
stemmer = PorterStemmer()
for word in words :
  print(f"{word} : {stemmer.stem(word)}")
```

```python
from nltk.stem import LancasterStemmer
Lancaster = LancasterStemmer()
for i in words:
  print(f"{i} : {Lancaster.stem(i)}")
```

# **RegexStemmer Class**
NLTK has RegexpStemmer class with the help of which we can easily implement Regular
Expression Stemmer algorithms. It basically takes a single regular expression and removes any
prefix or suffix that matches the expression. Let us see an example


```python
from nltk.stem import RegexpStemmer
regex = RegexpStemmer("ing|s$|e$|able$")
regex.stem("eating")
regex.stem("Cars")
```

```python
regex.stem("Understanding")
```

# **SnowBall Stemmer**

```python
from nltk.stem import SnowballStemmer
snow = SnowballStemmer("english")
def stemmer(model,list_words):
  for i in list_words:
    print(f"{i} : {model.stem(i)}")


stemmer(snow,words)
```

# **Lemitization**
Lemmatization technique is like stemming. The output we will after lemmatization is called
'lemma', which is a root word rather than root stem, the output of stemming. After lemmatization,
we will be getting a valid word that means the same thing.
NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This
class uses morphv() function to the WordNet CorpusReader class to find a lemma.

***Wornet Lemitizer***

```python
# import nltk
# nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_words(word_list,post_tag):
  for word in word_list:
    print(f"{word} : {lemmatizer.lemmatize(word,f"{post_tag}")}")

lemmatize_words(words,'v')
```
